```python
session_directory = 'sessions'

if os.path.exists(session_directory):
    shutil.rmtree(session_directory)
os.makedirs(session_directory)
os.makedirs("{}/csv".format(session_directory))
os.makedirs("{}/json".format(session_directory))
```

```python
for i in range(1,245):
    url_string = "http://api.offenesparlament.de/api/session/{}".format(i)
    with urllib.request.urlopen(url_string) as url:
        data = json.loads(url.read().decode())['data']
    df = pd.DataFrame(data)
    df['speaker_key'] = df['speaker_key'].apply(lambda x : '' if pd.isnull(x) else str(x).replace('.0', '') )
    df.to_csv("{0}/csv/{1:03d}.csv".format(session_directory, i, float_format='%.5f'))
    df.to_json("{0}/json/{1:03d}.json".format(session_directory, i), orient='records')
    
```
